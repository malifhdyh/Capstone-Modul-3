## **IQR (Interquartile Range)**

Untuk menentukan titik data mana yang dianggap sebagai outlier, kita dapat menggunakan ukuran statistik seperti **IQR (Interquartile Range)**. IQR dihitung sebagai selisih antara nilai persentil ke-75 (Q3) dan persentil ke-25 (Q1) dari suatu dataset. Titik data yang berada di bawah **Q1 - 1.5 × IQR** atau di atas **Q3 + 1.5 × IQR** dapat dianggap sebagai outlier.

```python
# Dictionary to store outliers' information
outliers_info = {}

# List of numerical columns
numerical_columns = ['Year', 'Engine_Size', 'Mileage', 'Price']

for column in numerical_columns:
    # Calculate Q1, Q3, and IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers_condition = (df[column] < lower_bound) | (df[column] > upper_bound)
    outliers = df[outliers_condition]

    # Store information in the dictionary
    outliers_info[column] = {
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'number_of_outliers': outliers_condition.sum(),  # Count of outliers
        'outliers_indices': outliers.index.tolist()  # Indices of outliers
    }


# summary for outliers_summary

outliers_summary = {key: {'bounds': (value['lower_bound'], value['upper_bound']), 
                          'number_of_outliers': value['number_of_outliers']} 
                    for key, value in outliers_info.items()}

# Creating DataFrame from 'outliers_summary'
outliers_summary_df = pd.DataFrame.from_dict(outliers_summary, orient='index')

# Splitting 'bounds' into 'lower_bound' and 'upper_bound' columns
outliers_summary_df[['lower_bound', 'upper_bound']] = outliers_summary_df['bounds'].apply(pd.Series)

# Dropping the original 'bounds' column
outliers_summary_df.drop('bounds', axis=1, inplace=True)

# Reordering the columns for a better view
outliers_summary_df = outliers_summary_df[['lower_bound', 'upper_bound', 'number_of_outliers']]

# Display DataFrame as a styled table
styled_table = outliers_summary_df.style.format('{:.0f}')
styled_table

```

1. **Year:** Terdapat 246 outlier. Mobil yang diproduksi sebelum tahun 2003 atau setelah tahun 2027 dianggap sebagai outlier.

2. **Engine Size:** Terdapat 23 outlier. Mobil dengan kapasitas mesin kurang dari -2 atau lebih besar dari 8 dianggap sebagai outlier.

# **Data Splitting**

---

Pembagian data (*data splitting*) adalah praktik membagi dataset menjadi dua atau lebih subset untuk memfasilitasi pelatihan, evaluasi, dan pengujian model. Pemisahan yang paling umum adalah antara *training set* dan *testing set*. *Training set* digunakan untuk melatih model, sedangkan *testing set* digunakan untuk mengevaluasi kinerja model pada data yang belum pernah dilihat sebelumnya.

Untuk dataset ini yang digunakan untuk memprediksi harga mobil bekas di Arab Saudi, rasio pemisahan yang umum digunakan adalah 80% untuk pelatihan dan 20% untuk pengujian. Artinya, kita menggunakan 80% data untuk melatih model *machine learning* dan 20% sisanya untuk mengevaluasi kinerjanya.

Berikut alasan umum penggunaan rasio 80-20:

1. **Training Set (80%):**

   * **Tujuan:** Digunakan untuk melatih model *machine learning*.
   * **Alasan:** Porsi yang lebih besar diberikan untuk pelatihan agar model dapat mempelajari pola dan hubungan pada data secara efektif.

2. **Testing Set (20%):**

   * **Tujuan:** Digunakan untuk mengevaluasi kinerja model pada data baru yang belum pernah dilihat.
   * **Alasan:** Porsi yang lebih kecil disisihkan untuk pengujian guna menilai seberapa baik model dapat melakukan generalisasi ke data baru.

Pemilihan rasio pembagian *training-testing* bergantung pada berbagai faktor, termasuk ukuran dataset dan kebutuhan spesifik dari tugas pemodelan. Namun, rasio 80-20 adalah titik awal yang umum digunakan, terutama ketika dataset berukuran sedang seperti 5.000 data.

**Dalam kasus ini, dengan 5.000 data, pembagian 80-20 berarti menggunakan 80% data (4.000 sampel) untuk melatih model *machine learning* dan menyisakan 20% data (1.000 sampel) untuk pengujian.**

1. **Dataset Kecil** (kurang dari 10.000 data):

   * Umum: 80-20 (80% untuk pelatihan, 20% untuk pengujian)
   * Alternatif: 70-30 (70% untuk pelatihan, 30% untuk pengujian)

2. **Dataset Sedang** (10.000 – 100.000 data):

   * Umum: 80-20
   * Alternatif: 75-25

3. **Dataset Besar** (lebih dari 100.000 data):

   * Umum: 80-20 atau 90-10
   * Alternatif: 85-15 atau 90-10

**6. Options:** Bertipe data *object* dengan 3 entri unik. ‘Semi Full’ adalah salah satu contoh entri uniknya.


```python
transformer = ColumnTransformer([
    ('One Hot', OneHotEncoder(drop='first'), ['Options', 'Origin', 'Gear_Type']),
    ('Binary', BinaryEncoder(), ['Type', 'Region', 'Make']),
    ('Robust', RobustScaler(),['Year','Engine_Size', 'Mileage'])
], remainder='passthrough')

transformer
```

**Penjelasan singkat tentang transformer kolom**

**1. OneHotEncoder:** Jenis pengkodean untuk fitur kategorikal. Membuat kolom baru yang menunjukkan keberadaan (atau ketidakberadaan) setiap nilai yang mungkin ada pada data asli. Argumen `drop='first'` berarti kategori pertama akan dihapus untuk menghindari *multicollinearity*, yaitu masalah umum ketika menggunakan variabel dummy.

**2. BinaryEncoder:** Jenis pengkodean lain untuk fitur kategorikal. Pertama mengubah kategori menjadi label numerik, lalu mengubah label tersebut menjadi kode biner. Setiap digit pada kode biner direpresentasikan sebagai kolom terpisah.

**3. RobustScaler:** Metode *feature scaling* yang menghapus nilai median dan menyesuaikan skala data berdasarkan *Interquartile Range* (IQR). Metode ini tahan terhadap *outlier*, sehingga cocok digunakan ketika data memiliki *outlier*.

**ColumnTransformer menerapkan transformasi ini pada kolom yang berbeda:**

1. **OneHotEncoder** diterapkan pada kolom ‘Gear\_Type’, ‘Origin’, dan ‘Options’.
2. **BinaryEncoder** diterapkan pada kolom ‘Type’, ‘Region’, dan ‘Make’.
3. **RobustScaler** diterapkan pada kolom ‘Year’, ‘Engine\_Size’, dan ‘Mileage’.
4. Argumen `remainder='passthrough'` berarti kolom lain yang tidak disebutkan akan diteruskan tanpa transformasi. Jika kolom tidak disebutkan di *transformers*, kolom tersebut akan dihapus.

**Alasan pemilihan metode:**

* **OneHotEncoder** untuk ‘Gear\_Type’, ‘Origin’, dan ‘Options’: Kolom ini memiliki sedikit kategori unik. One-hot encoding hanya akan menambah sedikit kolom baru dan tidak menambah dimensi data secara berlebihan. Karena sifatnya *nominal* (tidak memiliki urutan), metode ini tepat digunakan.
* **BinaryEncoder** untuk ‘Type’, ‘Region’, dan ‘Make’: Kolom ini, terutama ‘Type’ dan ‘Make’, memiliki jumlah kategori unik yang cukup banyak. Binary encoding lebih efisien dalam penggunaan ruang dibandingkan one-hot encoding dan membantu mengurangi dimensi data.
* **RobustScaler** untuk ‘Year’, ‘Engine\_Size’, dan ‘Mileage’: RobustScaler menggunakan statistik yang tahan terhadap *outlier*, berbeda dengan StandardScaler yang sensitif terhadap *outlier*. Oleh karena itu, metode ini tepat untuk variabel numerik yang mungkin memiliki *outlier*.

**Kesimpulan:**

Pemilihan *encoder* dan *scaler* didasarkan pada jumlah nilai unik serta sifat dari setiap kolom (kategorikal atau numerik, ada atau tidaknya *outlier*). RobustScaler dipilih karena sifatnya yang tahan terhadap *outlier* sehingga tidak sensitif terhadap nilai ekstrim.

## **Multicollinearity Test**
---

Multikolinearitas adalah fenomena di mana dua atau lebih fitur dalam model regresi memiliki korelasi tinggi satu sama lain. Hal ini dapat menimbulkan masalah dalam menginterpretasikan hasil model, karena menjadi sulit untuk menentukan kontribusi masing-masing fitur secara individual terhadap variabel target. Dalam konteks ini, kita perlu memeriksa apakah terdapat korelasi tinggi antar fitur di dataset yang dapat memengaruhi interpretasi model.

Untuk memeriksa multikolinearitas, kita dapat menggunakan *correlation matrix* pada fitur-fitur yang ada di dataset. Jika terdapat korelasi tinggi antara dua atau lebih fitur, itu menandakan adanya multikolinearitas.

Selain itu, kita juga dapat menggunakan **Variance Inflation Factor (VIF)** untuk mengukur tingkat multikolinearitas antar fitur. Jika nilai VIF suatu fitur melebihi ambang batas tertentu, hal itu mengindikasikan adanya multikolinearitas signifikan dengan fitur lain di model.

1. **Correlation Matrix:** Kita dapat membuat matriks korelasi pada dataset untuk mengidentifikasi adanya korelasi tinggi antar fitur. Jika ada pasangan fitur yang memiliki koefisien korelasi mendekati 1 atau -1, ini menunjukkan tingkat multikolinearitas yang tinggi.

2. **Variance Inflation Factor (VIF):** Setelah mengidentifikasi potensi multikolinearitas dari *correlation matrix*, kita dapat menghitung nilai VIF untuk setiap fitur guna mengukur tingkat keparahan multikolinearitas. Aturan praktis yang umum digunakan adalah jika nilai VIF lebih besar dari 4, maka multikolinearitas dianggap tinggi.

3. **Mengatasi Multikolinearitas:** Jika ditemukan adanya multikolinearitas, kita mungkin perlu menanganinya untuk meningkatkan kinerja dan interpretabilitas model. Penanganannya dapat meliputi menghapus salah satu fitur yang berkorelasi, menggabungkan fitur yang berkorelasi, atau menggunakan teknik regularisasi.


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python

# untuk menampilkan nilai VIF masing-masing feature
def calc_vif(X):
    vif = pd.DataFrame()
    
    vif['feature'] =  X.columns
    vif['VIF'] =[variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
    vif['acceptance'] = np.where(vif.VIF < 4, 'Yes', 'No')
    
    return vif
```

VIF untuk setiap variabel pada model ini adalah ukuran multikolinearitas, yang terjadi ketika fitur-fitur data tidak saling independen. Nilai VIF yang tinggi menunjukkan bahwa variabel tersebut sangat berkorelasi dengan variabel lain dalam model, yang dapat memengaruhi keandalan model ini.

> ## **Penjelasan Singkat untuk Metrik yang Dipilih**

**Root Mean Square Error (RMSE):**

* Kelebihan: Sensitif terhadap kesalahan yang besar dan dapat menunjukkan seberapa rapat data aktual mengelompok di sekitar nilai prediksi. Mudah diinterpretasikan karena satuannya sama dengan variabel target.
* Kekurangan: Sensitif terhadap *outlier* dan dapat memberikan hasil yang berlebihan jika terdapat *outlier* dalam dataset.

**Mean Absolute Error (MAE):**

* Kelebihan: Sederhana untuk dipahami dan dihitung. Hasilnya memiliki satuan yang sama dengan variabel output.
* Kekurangan: Tidak terdiferensiasi di semua titik, sehingga dapat menjadi masalah untuk algoritma optimasi tertentu.

**Mean Absolute Percentage Error (MAPE):**

* Kelebihan: Mudah dipahami karena kesalahan ditampilkan dalam bentuk persentase. Menguntungkan karena menunjukkan perbedaan dalam bentuk persen.
* Kekurangan: Tidak dapat digunakan jika nilai aktual mendekati atau sama dengan nol. Bersifat asimetris, artinya akan melaporkan kesalahan lebih tinggi jika nilai prediksi lebih besar dari nilai aktual, dan kesalahan lebih rendah jika prediksi lebih kecil dari nilai aktual.

**Mean Squared Error (MSE):**

* Kelebihan: Terdiferensiasi di semua titik, sehingga bermanfaat untuk algoritma optimasi selama pelatihan model.
* Kekurangan: Sangat sensitif terhadap *outlier* karena kesalahan dikuadratkan, sehingga dampaknya diperbesar terhadap metrik keseluruhan.

**Root Mean Squared Percentage Error (RMSPE):**

* Kelebihan: Mengukur rasio antara nilai prediksi dan nilai aktual. Cocok digunakan ketika target memiliki pertumbuhan eksponensial, seperti jumlah populasi atau rata-rata penjualan selama beberapa tahun.
* Kekurangan: Kekurangan spesifik tidak banyak dibahas dalam literatur, namun seperti RMSE, RMSPE juga dapat sensitif terhadap *outlier*.

**Root Mean Squared Logarithmic Error (RMSLE):**

* Kelebihan: Lebih tahan terhadap efek *outlier* dan bermanfaat jika tidak ingin memberikan penalti besar terhadap perbedaan besar antara prediksi dan nilai aktual ketika keduanya bernilai sangat besar.
* Kekurangan: Nilai error tidak diberikan dalam satuan variabel target, sehingga lebih sulit diinterpretasikan.

**R-Squared (R²):**

* Kelebihan: Menunjukkan proporsi variansi pada variabel dependen yang dapat dijelaskan oleh model. Skornya bebas skala, artinya meskipun nilai kecil atau besar, R² akan selalu berada di bawah satu.
* Kekurangan: Hanya dapat digunakan untuk mengevaluasi model linear. Juga mengasumsikan setiap variabel independen dalam model memiliki tingkat kepentingan yang sama.


**Untuk dataset ini saya menggunakan RMSE, MAE, dan MAPE untuk memprediksi harga mobil bekas di Arab Saudi.**

**1. Root Mean Square Error (RMSE):**
Metrik ini memberikan gambaran tentang seberapa besar kesalahan yang biasanya dibuat sistem dalam prediksinya, dengan memberikan bobot lebih besar pada kesalahan yang besar. RMSE adalah pilihan yang baik jika kesalahan besar sangat tidak diinginkan. RMSE merupakan metrik populer untuk masalah regresi seperti pada dataset ini. Metrik ini mengukur simpangan baku dari residual (*prediction errors*). Residual adalah ukuran seberapa jauh titik data dari garis regresi, dan RMSE mengukur seberapa menyebar residual tersebut. Dengan kata lain, RMSE menunjukkan seberapa rapat data terhadap garis *best fit*. Karena RMSE memberikan bobot lebih besar pada kesalahan besar, metrik ini lebih berguna ketika kesalahan besar sangat tidak diinginkan.

**2. Mean Absolute Error (MAE):**
MAE adalah rata-rata selisih absolut antara nilai prediksi dan nilai aktual. Metrik ini sederhana dan langsung yang dapat memberikan gambaran umum tentang kinerja model. MAE mengukur rata-rata besar kesalahan pada sekumpulan prediksi, tanpa mempertimbangkan arah kesalahan. Nilai MAE adalah rata-rata dari perbedaan absolut antara prediksi dan pengamatan aktual pada data uji, di mana semua perbedaan memiliki bobot yang sama. Karena itu, MAE memberikan ukuran yang jelas dan sederhana tentang seberapa jauh prediksi dari nilai sebenarnya.

**3. Mean Absolute Percentage Error (MAPE):**
MAPE menyatakan kesalahan prediksi dalam bentuk persentase sehingga mudah diinterpretasikan dan dipahami. MAPE mengukur kesalahan prediksi absolut dalam bentuk persentase terhadap nilai aktual. Hal ini sangat berguna jika ingin mengetahui seberapa besar kesalahan prediksi dibandingkan dengan nilai aktualnya.

Adapun penggunaan **R²**, metrik ini biasanya tidak digunakan pada model *gradient boosting* seperti CatBoost dan sejenisnya. R² adalah ukuran seberapa baik prediksi model cocok dengan data aktual, yang **lebih sesuai untuk model regresi linear.**

Sedangkan **RMSLE** dan **RMSPE** memang lebih berguna ketika variabel target (dalam hal ini harga) memiliki rentang nilai yang sangat luas. Jika rentang harga pada dataset ini relatif kecil, kedua metrik tersebut mungkin kurang informatif.

**MSE** memberikan penekanan lebih besar pada kesalahan yang besar karena perbedaan prediksi dan nilai aktual dikuadratkan. Dalam kasus seperti di PT AutoPredict Arabia, di mana kesalahan prediksi besar bisa berdampak signifikan, MSE mungkin kurang ideal karena dapat memberikan pengaruh berlebihan terhadap evaluasi keseluruhan. Walaupun MSE bisa baik di beberapa kasus, dalam konteks prediksi harga mobil bekas, MSE bisa saja kurang tepat karena sifatnya yang memperbesar dampak kesalahan besar.

**Kesimpulannya, pemilihan metrik harus selaras dengan tujuan bisnis dan karakteristik data. RMSE, MAE, dan MAPE memberikan sudut pandang yang berbeda terkait kinerja model, dan penggunaan ketiganya akan memberikan pemahaman yang lebih komprehensif untuk memprediksi harga mobil bekas di Arab Saudi.**

## **Penjelasan Hiperparameter CatBoost yang Digunakan untuk Mengatasi Masalah Overfitting**

```python
hyperparam_space_catboost = {
    'model__random_strength': [0.3, 0.5, 0.7, 1.0],  # Increased values
    'model__max_bin': list(range(2, 41, 2)),
    'model__bootstrap_type': ['Bayesian', 'Bernoulli', 'MVS', 'No'],
    'model__min_data_in_leaf': list(range(5, 31, 5)),  # Increased values
    'model__max_ctr_complexity': [1, 2, 3],  # Decreased range
    'model__learning_rate': [0.001, 0.01, 0.1],  # Reduced options
    'model__iterations': list(range(100, 401, 50)),
    'model__depth': list(range(5, 13)),
    'model__l2_leaf_reg': list(range(3, 9)),  # Increased lower bound
}
```

**1. random_strength:** Parameter ini menyeimbangkan pengaruh fitur kategorikal selama pemilihan split. Jika dalam dataset ini terdapat fitur kategorikal, menyesuaikan parameter ini dapat membantu meningkatkan performa model.

**2. max_bin:** Parameter ini menetapkan jumlah maksimum bin tempat nilai fitur akan dikelompokkan. Nilai yang lebih kecil dapat menyebabkan underfitting, sedangkan nilai yang lebih besar dapat menyebabkan overfitting.

**3. bootstrap_type:** Parameter ini menentukan metode pengambilan sampel dataset. Metode yang berbeda dapat menghasilkan performa model yang berbeda.

**4. min_data_in_leaf:** Parameter ini dapat membantu mencegah overfitting dengan menentukan jumlah minimal instance pelatihan dalam sebuah node daun. Nilai yang lebih besar dapat membantu mencegah overfitting tetapi mungkin menyebabkan underfitting.

**5. max_ctr_complexity:** Parameter ini menentukan jumlah maksimum fitur kategorikal yang dapat dikombinasikan. Nilai yang lebih besar dapat menangkap pola yang lebih kompleks tetapi dapat menyebabkan overfitting.

**6. learning_rate:** Parameter ini mengontrol ukuran langkah pada setiap iterasi saat bergerak menuju minimum fungsi kerugian. Learning rate yang lebih kecil dapat menghasilkan pembelajaran yang lebih akurat, tetapi proses pembelajarannya bisa lebih lambat.

**7. iterations:** Parameter ini menentukan jumlah pohon yang akan dibangun. Semakin banyak pohon dapat menangkap pola yang lebih kompleks tetapi dapat menyebabkan overfitting.

**8. depth:** Parameter ini menentukan kedalaman pohon. Pohon yang lebih dalam dapat menangkap pola yang lebih kompleks tetapi dapat menyebabkan overfitting.

**9. l2_leaf_reg:** Parameter ini mengontrol regularisasi L2 pada bobot, yang dapat membantu mencegah overfitting.

Kesimpulan:

Jika model CatBoost ini mengalami overfitting, kita dapat mempertimbangkan untuk menyesuaikan parameter berikut untuk mengatasinya:

- Meningkatkan random_strength.

- Meningkatkan min_data_in_leaf.

- Mengurangi max_ctr_complexity.

- Meningkatkan l2_leaf_reg.

```python
catboost = CatBoostRegressor(random_state=7, verbose=False)

crossval = KFold(n_splits=5, shuffle=True, random_state=7)

estimator_catboost = Pipeline([
    ('preprocessing', transformer),
    ('model', catboost)
])

rs = RandomizedSearchCV(
    estimator_catboost,
    param_distributions=hyperparam_space_catboost,
    cv=crossval,
    scoring=['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'],
    refit='neg_mean_absolute_error',
    n_jobs=-1,
    n_iter= 100,
    error_score='raise',
    random_state=7  
)
```

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming df_y_pred is already defined
df_y_pred = pd.DataFrame({'y_test': y_test, 'y_test_pred': y_pred_cat_tuning})
df_y_pred.reset_index(drop=True, inplace=True)
df_y_pred.sort_values('y_test', inplace=True)

def calculate_metrics(df):
    if len(df) == 0:
        return 0, 0, 0, 0  # Return default values when the subset is empty
    mae = mean_absolute_error(df['y_test'], df['y_test_pred'])
    rmse = np.sqrt(mean_squared_error(df['y_test'], df['y_test_pred']))
    mape = mean_absolute_percentage_error(df['y_test'], df['y_test_pred']) * 100
    length = len(df)
    return mae, rmse, mape, length

ranges = [(5000, 10000), (10000, 20000), (20000, 30000), (30000, 40000),
          (40000, 50000), (50000, 100000), (100000, 150000), (150000, 200000),
          (200000, 250000), (250000, 300000), (300000, 350000), (350000, 400000),
          (400000, 450000), (450000, 500000), (500000, float('inf'))]

list_range2 = ['{}-{}'.format(start, end) for start, end in ranges[:-1]] + ['>{}'.format(ranges[-1][0])]

list_metrics = []

for start, end in ranges:
    subset = df_y_pred[(df_y_pred['y_test'] >= start) & (df_y_pred['y_test'] < end)]
    mae, rmse, mape, length = calculate_metrics(subset)
    list_metrics.append((mae, rmse, mape, length))

columns = ["Score MAE (SAR)", "Score RMSE (SAR)", "Score MAPE (%)", "total Data"]
range_metrics = pd.DataFrame(list_metrics, columns=columns)
range_metrics["Range Price (SAR)"] = list_range2

# Reorder columns for better readability
range_metrics = range_metrics[["Range Price (SAR)", "Score MAE (SAR)", "Score RMSE (SAR)", "Score MAPE (%)", "total Data"]]

range_metrics
```

   * **Mid** (20k–150k SAR): model utama (volume terbesar).
   * **High** (>150k SAR): optimalkan **MAE/RMSE**, pertimbangkan **log-price** target, **sample weighting ∝ price**,/atau **Quantile/Pinball loss** untuk menekan underpricing.
3. **Transformasi target & pembobotan:** uji **log(Price)** untuk menstabilkan varians; terapkan **price-weighted loss** agar mobil mahal diberi bobot lebih besar pada training.
4. **Kalibrasi & interval harga:** berikan **rentang rekomendasi (P10–P90)** agar tim pricing punya guard-rail (hindari over/underpricing ekstrem).
5. **Penanganan ketimpangan data:** lakukan **binning permintaan** per rentang harga + **stratified sampling** saat training; pertimbangkan **data augmentation** (synthetic combos yang masuk akal) khusus high-end bila memungkinkan.
6. **Feature engineering prioritas:**

   * Kondisi unit (grade inspeksi, hasil checklist, cat ulang/karat), **riwayat servis & klaim asuransi**, jumlah kepemilikan, **days-on-market**, jumlah view/lead, reputasi dealer, musiman/kalender, **kelangkaan varian**.
   * Variabel regional (urban vs non-urban, indeks permintaan lokal), harga BBM, dan **trend likuiditas per merek/seri**.
7. **Monitoring & retraining:** track **MAPE per rentang** + **MAE/RMSE untuk high-end** mingguan; siapkan **drift alerts** (data/target) dan **retraining** berkala (mis. bulanan) atau saat metrik melewati ambang.

### Rekomendasi Operasional

1. **Guard-rails penetapan harga:** batasi perubahan harga otomatis (mis. ±5–8% dari harga acuan internal). Kasus di luar rentang → **manual review**.
2. **A/B test bertahap:** uji harga rekomendasi pada sebagian inventori/region. Key metrics: **gross margin/unit**, **sell-through rate**, **time-to-sale**, dan **win-rate vs kompetitor**.
3. **Dashboard operasional:** metrik harian per segmen harga/merek/region, heatmap error, outlier watchlist (unit yang jauh dari rentang P10–P90).
4. **SOP akuisisi data:** integrasi alur **inspeksi standar** agar fitur kondisi selalu terekam; pastikan data “opsi” (trim/paket) seragam penamaannya (kurangi kebisingan kategori).
5. **Governance & explainability:** tampilkan **alasan harga** (top-features untuk unit tertentu) agar tim sales/buyer dapat menjelaskan ke seller/pembeli.


## **BISNIS**


### Ringkasan Eksekutif

* **Sweet spot harga:** inventori paling likuid dan prediksi paling akurat berada di **100k–250k SAR**; segmen ini seharusnya jadi **core business**.
* **Ekstrem harga tidak stabil:** <20k SAR dan >350k SAR berisiko tinggi (data tipis, varian besar). Perlakukan dengan kebijakan khusus (clearance & manual review).
* **Uplift potensial:** simulasi harga rekomendasi menunjukkan potensi **+1,35%** pendapatan (\~**+1,15M SAR**) jika diterapkan konsisten pada seluruh unit terjual.

### Wawasan Pasar & Portofolio

* **Distribusi saat ini timpang** ke 50k–150k SAR (volume tinggi), yang berarti **permintaan mass-market** kuat.
* **Year & Engine Size** paling memengaruhi harga → stok lebih muda & mesin ideal (bukan terbesar) cenderung **lebih cepat laku**.
* **High-end** (luxury/varian langka) butuh **storytelling & verifikasi kondisi** untuk menjustifikasi harga (bukan sekadar diskon).

### Strategi Harga (Revenue & Margin)

1. **Dynamic pricing dengan guard-rails.** Terapkan harga rekomendasi model dengan **batas ±5–8%** dari acuan internal. Di luar batas → **manual approval**.
2. **Segmentasi kebijakan harga:**

   * **Low** (<20k): strategi **clearance cepat**; markup tipis, rotasi stok.
   * **Mid** (20k–150k & 150k–250k): **standar margin**, update harga mingguan berdasar pergerakan kompetitor & days-on-market.
   * **High** (>250k): fokus **MAE absolut** (hindari underpricing besar), sertakan **rentang harga (P10–P90)** di listing agar negosiasi terarah.
3. **Markdown berbasis usia stok:**

   * **Day 0–30:** harga rekomendasi + full margin.
   * **Day 31–60:** diskon **1–2%** bertahap.
   * **>60 hari:** eskalasi ke **program promo/financing** atau remarketing lintas region.
4. **Add-on monetization:** bundel **garansi, asuransi, detailing, dan servis awal**; target **attach-rate 35–45%** di mid-segment.

### Strategi Inventory & Procurement

1. **Double-down mid-segment:** tingkatkan porsi stok **100k–250k SAR** (paling likuid) hingga **+15–25%** dari baseline.
2. **Kurangi ekor ekstrem:** batasi akuisisi unit **<10k** dan **>350k** ke **<5%** total stok kecuali ada permintaan PO/booking.
3. **Playbook pembelian:** gunakan **harga rekomendasi beli** (buy-box) dengan target **gross margin per unit** minimal **x%** (tetapkan angka internal), beda per segmen.
4. **Kemitraan pasokan:** kerja sama dealer/pelelangan untuk **varian cepat laku**; kontrak **buy-back** terbatas untuk luxury agar risiko harga turun tertahan.

### Go-To-Market & Penjualan

* **Listing excellence:** foto konsisten (studio-like), 360°, riwayat servis, nilai inspeksi → tingkatkan CTR & trust.
* **Omnichannel:** sinkronkan marketplace, situs sendiri, showroom; **lead-routing SLA <10 menit**.
* **Pembiayaan & promo:** kolaborasi bank/leasing; **DP ringan** untuk mid-segment; **buyback guarantee** terbatas untuk luxury.
* **Kampanye tematik:** “**Certified & Inspected**”, “**0 Claim History**”, dan “**Best Value 150k–200k**”.

### Operasi & Layanan

* **Standarisasi inspeksi & grading** (kelistrikan, body, interior, ban, rem). Tampilkan **skor kondisi** di listing.
* **SLA reconditioning**: ≤**5 hari** dari akuisisi ke siap jual; dashboard bottleneck bengkel.
* **After-sales:** garansi powertrain 3–6 bulan untuk mid-segment → naikkan konversi & NPS.

### KPI & Target 90 Hari

* **Unit Economics:** Margin/unit, **gross profit per unit (GPU)**, attach-rate add-on.
* **Kecepatan:** **Days-on-Market (DOM)** median; target **-15%** di segmen mid.
* **Harga:** **Revenue Uplift vs Harga Aktual** (target **≥ +1%**) & **Leakage** (selisih vs rentang P10–P90).
* **Penjualan:** Conversion rate lead→deal, waktu respons lead (<10 menit), NPS/Rating.
* **Stok:** rotasi stok (turns) per segmen, % aged stock >60 hari (<12%).

### Risiko Utama & Mitigasi

* **Sparsity segmen ekstrem:** batasi eksposur, pakai **manual review** & storytelling nilai.
* **Data kualitas & inkonsistensi opsi/trim:** standardisasi penamaan & audit mingguan.
* **Reaksi kompetitor:** gunakan **rentang harga** + benefit non-harga (garansi, servis, pembiayaan).
* **Underpricing high-end:** gunakan **MAE absolut** & **price floor** berbasis pembobotan harga.
